```python
# dependecies
import pandas as pd
from tabgan.sampler import GANGenerator
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('personalFinanceDataset.csv')
```

```python
data = data.drop(['Unnamed: 0'], axis=1)
data = data.drop(['medicalDebt'], axis=1)
data = data.drop(['gaming'], axis=1)
data.head()
```

```python
targetColumn = 'financialHealth'
x = data.drop(columns=[targetColumn])
featureNames = x.columns
y = data[targetColumn]
```

```python
catCols = data.select_dtypes(include=['object', 'boolean']).columns.tolist()
numCols = data.select_dtypes(include=['number']).columns.tolist()
```

```python
scaler = StandardScaler()
```

```python
# split
xtrain, xval, ytrain, yval = train_test_split(x, y, test_size=0.2, random_state=42)
xtrain[numCols] = scaler.fit_transform(xtrain[numCols])
xval[numCols] = scaler.transform(xval[numCols])
```

```python
gan = GANGenerator(
    gen_params = {'batch_size': 500, 'epochs': 300, 'patience': 10},
    cat_cols = catCols
).generate_data_pipe(
    train_df=xtrain,
    target=ytrain.to_frame(targetColumn),
    test_df=xval,
    deep_copy=True,
    only_generated_data=True
)
```
